# Weather Forecast

## Topic

In this notebook, I'll be dealing with a Regression problem with the aim to predict a continuous value. The weather.csv file contains a  dataset about different measured and recorded aspects of the weather at each day for almost a year. The entire dataset contains 366 entries and 22 columns which are:

- MinTemp : minimun temperature recorded that day (Target variable)
- MaxTemp : maximum temperature recorded that day
- Rainfall : amount of rainfall    
- Evaporation : amount of evoporation 
- Sunshine  : amount of sunshine   
- WindGustDir : direction of wind gust
- WindGustSpeed  : direction of wind gust
- WindDir9am  : direction of the wind at 9am
- WindDir3pm  : direction of the wind at 3pm   
- WindSpeed9am  : recorded speed of the wind at 9am
- WindSpeed3pm  : recorded speed of the wind at 3pm
- Humidity9am : recorded humidity at 9am   
- Humidity3pm  : recorded humidity at 3pm
- Pressure9am  : recorded pressure at 9am 
- Pressure3pm : recorded pressure at 3pm   
- Cloud9am  : amount of coulds at 9am     
- Cloud3pm  : amount of clouds at 3pm   
- Temp9am  : recorded temperature at 9am     
- Temp3pm  : recorded temperature at 3pm      
- RainToday : whether or not it rains today     
- RISK_MM : amount of rainfall in millimeters for the next day      
- RainTomorrow  : whether or not it rains tomorrow 

## Objective

Predict the minimum temperature of the day with the least error possible. This is a supervised machine learning task of regression.

### Split train and test sets

```python
train_set, test_set = train_test_split(weather_df, test_size = 0.2,random_state = 42)
```

```python
target_train = train_set["MinTemp"].copy()
train_set = train_set.drop(["MinTemp"], axis = 1)
target_test = test_set["MinTemp"].copy()
test_set = test_set.drop(["MinTemp"], axis = 1)
```

At this stage , I splited the dataframe into a train ans test set (80%, 20%)

### Prepare data for ML

```python
le_train = train_set[["RainToday","RainTomorrow"]].copy()
le_test = test_set[["RainToday","RainTomorrow"]].copy()
```

```python
onehot_train = train_set[["WindGustDir","WindDir9am","WindDir3pm"]].copy()
onehot_test = test_set[["WindGustDir","WindDir9am","WindDir3pm"]].copy()
```

```python
num_train = train_set.drop(["RainToday","RainTomorrow","WindGustDir","WindDir9am","WindDir3pm"], axis = 1)
num_test = test_set.drop(["RainToday","RainTomorrow","WindGustDir","WindDir9am","WindDir3pm"], axis = 1)
```

Now, since the dataset contains numerical values, categorical values with only two categories and categorical values with multiple categories, and since thesde wpn't be transformed in the same way, I went ahead and seperated them into 3 datsets.

```python
imputer = SimpleImputer(strategy='median')
imputer.fit(num_train)
num_train = imputer.transform(num_train)
num_test = imputer.transform(num_test)
num_train = pd.DataFrame(data = num_train, columns = ['MaxTemp', 'Rainfall', 'Evaporation', 'Sunshine', 'WindGustSpeed',
       'WindSpeed9am', 'WindSpeed3pm', 'Humidity9am', 'Humidity3pm',
       'Pressure9am', 'Pressure3pm', 'Cloud9am', 'Cloud3pm', 'Temp9am',
       'Temp3pm', 'RISK_MM'])
num_test = pd.DataFrame(data = num_test, columns = ['MaxTemp', 'Rainfall', 'Evaporation', 'Sunshine', 'WindGustSpeed',
       'WindSpeed9am', 'WindSpeed3pm', 'Humidity9am', 'Humidity3pm',
       'Pressure9am', 'Pressure3pm', 'Cloud9am', 'Cloud3pm', 'Temp9am',
       'Temp3pm', 'RISK_MM'])
```

I used simple imputer to fill the missing numerical values with the median of their column, then I transformed the result back into a dataframe (because it was transformed to an array by the imputer)

```python
imputer_cat = SimpleImputer(strategy='most_frequent')
imputer_cat.fit(le_train)
le_train = imputer_cat.transform(le_train)
le_test = imputer_cat.transform(le_test)
le_train = pd.DataFrame(data = le_train, columns = ["RainToday","RainTomorrow"])
le_test = pd.DataFrame(data = le_test, columns = ["RainToday","RainTomorrow"])
```

```python
le = LabelEncoder()
le_train["RainToday"] = le.fit_transform(le_train["RainToday"])
le_test["RainToday"] = le.fit_transform(le_test["RainToday"])
le_train["RainTomorrow"] = le.fit_transform(le_train["RainTomorrow"])
le_test["RainTomorrow"] = le.fit_transform(le_test["RainTomorrow"])
```

After that, I imputed the two-catgory categorical features with the most frequent category and then applied label encoding to the concerned columns 

```python
imputer_cat.fit(onehot_train)
onehot_train = imputer_cat.transform(onehot_train)
onehot_test = imputer_cat.transform(onehot_test)
onehot_train = pd.DataFrame(data = onehot_train, columns = ["WindGustDir","WindDir9am","WindDir3pm"])
onehot_test = pd.DataFrame(data = onehot_test, columns = ["WindGustDir","WindDir9am","WindDir3pm"])
```

```python
lb = LabelBinarizer()
onehot_train["WindGustDir"] = lb.fit_transform(onehot_train[["WindGustDir"]])
onehot_test["WindGustDir"] = lb.fit_transform(onehot_test[["WindGustDir"]])
onehot_train["WindDir9am"] = lb.fit_transform(onehot_train[["WindDir9am"]])
onehot_test["WindDir9am"] = lb.fit_transform(onehot_test[["WindDir9am"]])
onehot_train["WindDir3pm"] = lb.fit_transform(onehot_train[["WindDir3pm"]])
onehot_test["WindDir3pm"] = lb.fit_transform(onehot_test[["WindDir3pm"]])
```

Then, for the multiple categorical columns I used the same strategy for the imputer. And I encoded them using the LabelBinatizer which transforms each category to a number made of 0s and 1s

```python
train_pre = num_train.join(le_train)
train_pre = train_pre.join(onehot_train)
test_pre = num_test.join(le_test)
test_pre = test_pre.join(onehot_test)
```

Finally, I gethered the three datasets into one

```python
sts = StandardScaler()
X_train = sts.fit_transform(train_pre)
y_train = target_train
X_test = sts.fit_transform(test_pre)
y_test = target_test
```

And used the standard scaler (which is more efficient that the MinMax scaler at dealing with outliers)

## Apply ML algorithms

### Linear Regression

```python
lin_reg = LinearRegression()
lin_reg.fit(X_train,y_train)
```

```python
lin_reg_pred = lin_reg.predict(X_test)
lin_mse = mean_squared_error(y_test,lin_reg_pred)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```python
lin_abs = mean_absolute_error(y_test,lin_reg_pred)
lin_abs
```

This is the baseline model, although its results are not bad at all; it returns a root mean squared error of 2.22 and a mean absolute error of only 1.76

### Decision Tree Regressor

```python
decision_tree = DecisionTreeRegressor()
```

```python
parameters={"splitter":["best"],
            "max_depth" : [5,7,9],
           "min_samples_leaf":[1,2],
           "min_weight_fraction_leaf":[0.1,0.2],
           "max_features":["auto","log2"],
           "max_leaf_nodes":[None,10,20] }
dt_grid = GridSearchCV(decision_tree,param_grid = parameters,scoring='neg_mean_squared_error',cv=5 ,verbose=3)
dt_grid.fit(X_train, y_train)
```

```python
dt_grid.best_params_
```

```python
decision_tree = DecisionTreeRegressor(max_depth= 5,
 max_features= 'auto',
 max_leaf_nodes= 20,
 min_samples_leaf= 1,
 min_weight_fraction_leaf= 0.1,
 splitter= 'best')
decision_tree.fit(X_train, y_train)
```

```python
dec_pred = decision_tree.predict(X_test)
dec_mse = mean_squared_error(y_test,dec_pred)
dec_rmse = np.sqrt(dec_mse)
dec_rmse
```

```python
dec_abs = mean_absolute_error(y_test,dec_pred)
dec_abs
```

The decision tree regressor didn't show any better results than the baseline model, eventhough I used the gridsearch to fine tune its hyperparameters and used its best estimator to make the predictions. 

### Random Forest

```python
param_grid = [
    
    {'n_estimators' : [3 , 10 , 30], 'max_features' : [2 , 4 , 6 , 8]},
    {'bootstrap':[False] ,'n_estimators' : [3 , 10 ] , "max_features" : [2 ,3 ,4 ]},
]
```

```python
forest_reg = RandomForestRegressor()
grid_search = GridSearchCV(forest_reg , param_grid , cv = 5 , scoring = 'neg_mean_squared_error' , return_train_score= False)
grid_search.fit(X_train, y_train)
```

```python
best_ran = grid_search.best_estimator_
best_ran
```

```python
for_pred = best_ran.predict(X_test)
for_mse = mean_squared_error(y_test,for_pred)
for_rmse = np.sqrt(for_mse)
for_rmse
```

```python
for_abs = mean_absolute_error(y_test,for_pred)
for_abs
```

The random forest regressor performed slightly better that the decision tree, although its performance is still lower than that of the baseline.

### XGBoost Regressor

```python
xgb = XGBRegressor()
parameters = {'nthread':[4], 
              'objective':['reg:linear'],
              'learning_rate': [.03, 0.05, .07], 
              'max_depth': [5, 6, 7],
              'min_child_weight': [4],
              'silent': [1],
              'subsample': [0.7],
              'colsample_bytree': [0.7],
              'n_estimators': [500]}
```

```python
xgb_grid = GridSearchCV(xgb,
                        parameters,
                        cv = 2,
                        n_jobs = 5,
                        verbose=True)
xgb_grid.fit(X_train, y_train)
```

```python
xgb_grid.best_params_
```

```python
xgb_pred = xgb_grid.predict(X_test)
xgb_mse = mean_squared_error(y_test,xgb_pred)
xgb_rmse = np.sqrt(xgb_mse)
xgb_rmse
```

```python
xgb_abs = mean_absolute_error(y_test,xgb_pred)
xgb_abs
```

With XGB Regressor, I used the gridsearch once again to determine its best hyperparameters for this dataset. After making the predictions with it, we see that it made more errors than the baseline model but it's still more improved that the decision tree and random forest. 

### SVM Regressor

```python
param = {'kernel' : ('linear', 'poly', 'rbf', 'sigmoid'),'C' : [1,5,10],'degree' : [3,8],
         'coef0' : [0.01,10,0.5],'gamma' : ('auto','scale')}
svr = SVR()
svr_grid = GridSearchCV(svr, param, cv=5, n_jobs = -1, verbose = 2)
svr_grid.fit(X_train,y_train)
```

```python
svr_best = svr_grid.best_estimator_
svr_best
```

```python
svr_pred = svr_best.predict(X_test)
svr_mse = mean_squared_error(y_test,svr_pred)
svr_rmse = np.sqrt(svr_mse)
svr_rmse
```

```python
svr_abs = mean_absolute_error(y_test,svr_pred)
svr_abs
```

SVM Regressor brought about the best results of all, its predictions have very low error rates adn it outperformed the previous models.

## Comapring algorithms performance

```python
scores = [["LR", lin_mse ,lin_rmse, lin_abs],["DT", dec_mse, dec_rmse, dec_abs],
         ["RF", for_mse, for_rmse, for_abs], ["XGB", xgb_mse, xgb_rmse, xgb_abs],["SVR", svr_mse, svr_rmse,svr_abs]]
scores_df = pd.DataFrame(scores, columns = ["Model", "MSE", "RMSE", "MAE"])
scores_df = scores_df.sort_values(by = "RMSE",ascending = "False")
scores_df
```

```python
sns.catplot(data=scores_df, kind="bar",x ="Model", y="RMSE")
```

```python
sns.catplot(data=scores_df, kind="bar",x ="Model", y="MAE")
```

From the bar graphs, we see clearly that SVR scored the best with the least errors. I think the reason why Decision tree, random forest and XGB didn't work very well eventhough they are highly optimized algorithms, is that the relationships between most variables in this dataset are linear ones (as shown by the graphs before). Therefore , linear regression and SVR which based on finsing linear relationships between the instances worked the best here.

## Conclusion

Finally, we can say that SVR would be the algo used to best predict this dataset. The reason might be that the relationps between features are mostly linear.
